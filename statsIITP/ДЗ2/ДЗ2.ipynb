$$F(a, b) = \sum\limits_{i=0}^{n} (y_i - ax_i - b)^2$$

Оценим параметр $\sigma$,  Так как у нас есть наблюдения в одно  точке то мы можем взять разность этих наблюдений тем самым получит случайную  величину $\epsilon_i - \epsilon_{i+1}$, которая будет распределена нормально с параметрами $N(0, 2\sigma^2)$,  так как изначальные величины центрированные.

$$\epsilon_i - \epsilon_{i+1} = \xi_i \sim N(0, 2\sigma^2)$$
Дисперсия оценивается следующим образом
$$2\hat{\sigma}^2 = \frac{1}{n-1}\sum(\xi_i -\overline{\xi} )^2$$
$$\hat{\sigma} = \sqrt{\frac{1}{2(n-1)}\sum(\xi_i -\overline{\xi} )^2}$$

Найдем оценку для $a$ и $b$.

$$\frac{\partial F(a,b)} {\partial b} = 0$$
$$\sum \limits_{i=0}^{n} (y_i - ax_i - b) = 0 \rightarrow b = \overline{y} - a \overline{x}$$
$$\frac{\partial{F(a,b)}}{{\partial a}} = 0$$
$$\sum\limits_{i=0}^{n} x_i(y_i - ax_i - b) = 0 \rightarrow a = \frac{\overline{xy}- \overline{x}*\overline{y}}{\overline{x^2}- \overline{x}^2}$$

###### Задача 3: Гребневая регрессия
Дана гребневая регрессия (Ridge regression) может быть рассмотрена с точки зрения Байесовского подхода. В этой задаче мы исследуем влияние априориной ковариации на поведение модели. Пусть $\Sigma$ -корреляционная матрица d × d, соответствующая модели $Y = X \beta + \epsilon$ и оптимизационной задаче:
$$
RSS(\lambda) = ||y - X \beta||^2 + \lambda \beta^T \Sigma^{-1} \beta
$$

где $X$ - это матрица наблюдений n × d, а y - целевая переменная.
 

Четным номерам списка соответствует ниже приведенная матрица $\Sigma_1$, а нечетным - $\Sigma_2$;
Исследуйте влияние $\rho$ на оценки параметров модели. Как это соотноситься с  $\lambda$?

$$
\Sigma_1 = \begin{pmatrix}
 1 & \rho & \rho^2 & \dots & \rho^{d-1}\\
 \rho & 1 & \rho & \dots & \rho^{d-1}\\
 \vdots &   \vdots & \vdots &  & \vdots\\
\rho^{d-1} & \rho^{d-2} & \dots & \dots & 1
\end{pmatrix}
$$

$$
\Sigma_2 = \begin{pmatrix}
 1 & \rho & \rho & \dots & \rho\\
 \rho & 1 & \rho & \dots & \rho\\
 \vdots &   \vdots & \vdots &  & \vdots\\
\rho & \rho & \dots & \dots & 1
\end{pmatrix}
$$


$$RSS(\lambda) = ||y - X \beta||^2 + \lambda \beta^T \Sigma^{-1}\beta = (y - X \beta)^T(y - X \beta) + \lambda \beta^T \Sigma^{-1}\beta $$

$$\frac{\partial RSS}{\beta}  = -2X^Ty + 2X^TX\beta + 2\lambda \Sigma^{-1}\beta = 0$$
Мы воспользовались тем, что $\Sigma$ - симметричная значит и $\Sigma^{-1}$ тоже симметричная
$$\beta = (X^TX + \lambda \Sigma^{-1})^{-1}X^Ty$$

```python
from sklearn.datasets import make_regression
n = 1000
d = 100
sigma = 1
X, y = make_regression(n_samples=n, n_targets=1, n_features=d, n_informative=20, noise=sigma, shuffle=True, random_state=32)
split = int(X.shape[0] *0.8)
X_train,y_train = X[:split, :],y[:split]
X_test,y_test = X[split:, :], y[split:]
```

```python
def calcBeta(X, Z, lambda_,y):
    return (np.linalg.inv((X.T).dot(X) + lambda_*Z)).dot(X.T).dot(y)
```

```python
def createZ(rho,d):
    Z = np.ones((d,d)) * rho
    for i in range(d):
        Z[i][i] = 1
    return np.linalg.inv(Z)
```

```python
n = 10
m = 10
lambda_s = np.linspace(0,0.4,n)
rhos = np.linspace(0,2,m)
Mse = np.zeros((n,m))
betas = np.zeros((n,m))
for i, lambda_ in enumerate(lambda_s):
    for j,rho in enumerate(rhos):
        beta = calcBeta(X, createZ(rho, d), lambda_, y)
        Mse[i][j] = ((y_test - X_test.dot(beta))**2).sum()/y_test.shape[0]
        betas[i][j] = np.sqrt((beta **2).sum())
```

```python
plt.subplots(figsize=(20, 10))
for i in range(Mse.shape[0]):
    plt.plot(rhos, Mse[i,:], label=r"$\lambda = $" + str(lambda_s[i]));
plt.grid()
plt.ylabel("MSE")
plt.xlabel(r"$\rho$");
plt.legend()
# plt.yscale("log")
# plt.xscale("log")
```

Пик появляется при $\rho \rightarrow 1$ тогда матрица весов вырождена

Построим класическую картинку для ридж-регрессии,  зависимости MSE от $\lambda$.

Введем следующую матрицу
$$W_{\lambda} = (X^TX + \lambda Z^{-1})^{-1}X^TX$$
$$\hat{\beta} = (X^TX)^{-1}X^TY$$
$$\beta = (X^TX + \lambda \Sigma^{-1})^{-1}X^Ty$$

$$MSE[\hat{\beta(\lambda)}] = E[(W_{\lambda}\hat{\beta} - \beta)^T(W_{\lambda}\hat{\beta} - \beta)] = ... = \sigma^2tr\{W_{\lambda}(XX^T)^{-1} W_{\lambda}^T\} + \beta^T(W_{\lambda} - I)^T(W_{\lambda} - I)\beta$$

В общем случае
$$MSE(\theta) = E[(\hat{\theta} - \theta)^2] = Var(\hat{\theta}) + [Biass(\hat{\theta})]^2$$

Отсюда находим Var и Biass эстиматора, взято отсюда
https://arxiv.org/pdf/1509.09169;Lecture

```python
n =10
m = 10
Var = np.zeros((n,m))
Biass = np.zeros((n,m))
Mse = np.zeros((n,m))
lambda_s = np.linspace(0,2,n)
rhos = np.linspace(0,2,m)
```

```python
beta = calcBeta(X, createZ(0, d), 0, y) 
for i, lambda_ in enumerate(lambda_s):
    for j,rho in enumerate(rhos):
        W = calcBeta(X, createZ(rho, d), lambda_, X) 
        Biass[i][j] = np.sqrt(beta.T.dot((W - np.eye(W.shape[0], dtype=int)).T).dot((W - np.eye(W.shape[0], dtype=int))).dot(beta))
        Var[i][j] = sigma* np.trace(W.dot(np.linalg.inv(X.T.dot(X))).dot(W.T))
        Mse[i][j] = Var[i][j] + Biass[i][j]**2
```

###### Задача 4: Elasticnet 
Рассмотрим задачу регрессии $(x_i \in R^{d}, y_i \in R,$ $i=1,\dots,n)$ с двумя регуляризациями:
$$
RSS(\lambda, \mu) = ||y - X \beta||^2 + \lambda||\beta||^2_2 + \mu |\beta|_1 
$$
где $\lambda$ соответствует $L_2$-регуляризации, а $\mu$ - $L_1$-регуляризации.

1. В предположении, что $X^T X = I_d$, найдите значение $\beta^{elastic}$ в терминах $\beta^{ls}$, $\beta^{ls}$ соответствует значением параметров $\lambda = \mu = 0$.
2. Изобразите значение коэффициентов $\beta^{elastic}$ относительно $\beta^{ls}$.
3. Для данных из предыдущей задачи найдите оптимальные значения $\mu, \lambda$.

$$RSS(0,0)=||y - X \beta||^2 = (y - X\beta)^T(y-X\beta) $$
$$\frac{\partial RSS}{\partial \beta} = 0 = -2*X^Ty + 2X^TX\beta \rightarrow \beta^{ls} = X^Ty$$


$$RSS(0,0)=||y - X \beta||^2 + \lambda||\beta||_2^2 + \mu|\beta|_1 = (y - X\beta)^T(y-X\beta) + \lambda \beta^T\beta + \mu|\beta|$$
Попробуем минимизировать эту сумму. Мы можем выкидывать члены, котороые не влияют на $RSS$, тогда получим следующий функционал на минимизацию
$$-2y^TX\beta + \beta^T\beta + \lambda \beta^T\beta + \mu|\beta| \rightarrow min$$

$$L = -2\beta^{ls}\beta + \beta^T\beta + \lambda \beta^T\beta + \mu|\beta| \rightarrow min$$
В координатной форме:
$$L_i = -2\beta^{ls}_i\beta_i + \beta_i^2 + \lambda \beta^2_i + \mu|\beta_i| \rightarrow min$$
Рассмотрим 2 случая. Первый -  $\beta^{ls} > 0$, ясно, что тогда имеет смысл рассматривать $\beta_i \geq 0$. Тогда раскроем наши выражения:
$$L_i = -2\beta^{ls}_i\beta_i + \beta_i^2 + \lambda \beta^2_i + \mu\beta_i \rightarrow min$$
Дифференцируя по $\beta_i$ и приравнивая к 0.
$$\frac{\partial L_i}{\partial\beta_i} = -2\beta^{ls}_i + 2\beta_i + 2\lambda\beta_i + \mu = 0$$
$$\frac{\partial L_i}{\partial\beta_i} = -2\beta^{ls}_i + 2\beta_i + 2\lambda\beta_i + \mu = 0$$
$$\beta^{elastic}_i = \frac{(\beta^{ls} - \frac{\mu}{2})^+}{1+\lambda}$$
Второй  -  $\beta^{ls} \leq 0$, ясно, что имеет смысл рассматривать $\beta_i \leq 0$
$$L_i = -2\beta^{ls}_i\beta_i + \beta_i^2 + \lambda \beta^2_i + -\mu\beta_i \rightarrow min$$
$$\beta^{elastic}_i = \frac{(\beta^{ls} + \frac{\mu}{2})^-}{1+\lambda}$$
Комбинируя два результата получаем:
$$\beta^{elastic}_i = \frac{sign(\beta^{ls})(|\beta^{ls} | - \frac{\mu}{2})^+}{1+\lambda}$$

```python
plots_count = 5
lambdas = np.linspace(0,5,plots_count)
mus = np.linspace(0,5,plots_count)
n = 100
b_ls = np.linspace(-10,10, n)
b_elas = np.zeros(n)
```

```python

for lambda_, mu in zip(lambdas, mus):
    for i,val in enumerate(b_ls):
        b_elas[i] = np.sign(val)*(np.abs(val) - mu/2)/(1+lambda_) if (np.abs(val) - mu/2) > 0 else 0
    plt.plot(b_ls, b_elas, label="$\lambda = $"+str(lambda_) + "  $\mu = $"+str(mu))
plt.grid()
plt.xlabel(r"$\beta^{ls}$")
plt.ylabel(r"$\beta^{elastic}$")
plt.legend();
```

 $$RSS = (y - X\beta)^T(y-X\beta) + \lambda \beta^T\beta + \mu|\beta| $$
 $$\frac{\partial RSS}{\partial \beta} =   -2X^Ty + 2X^TX\beta + 2\lambda \beta + \mu sign(\beta)$$
 
  $$\frac{\partial RSS}{\partial \beta} = -2\beta^{ls} + 2X^TX\beta + 2\lambda \beta+  \mu sign(\beta)$$

```python
def getGradient(b_ls, b_elas, X, lambda_, mu_):
    return -2 * b_ls + 2*X.T.dot(X).dot(b_elas) + 2*lambda_ * b_elas + mu_*np.sign(b_elas)
```

```python
def calcBetaElastic(b_ls, b_elas, X ,lambda_, mu_, maxiter=1000):
    gradient_prev = np.zeros(b_ls.shape)
    b_elas = b_ls.copy()
    epsiiilon = 1e-2
    for k in range(1,maxiter):
        if(np.sum(getGradient(b_ls, b_elas, X, lambda_, mu_) ** 2) < epsiiilon):
            break
        gradient = getGradient(b_ls, b_elas, X, lambda_, mu_)
        if ((gradient - gradient_prev)**2).sum() < epsiiilon:
            break
        gradient_prev = gradient
        b_elas = b_elas - epsiiilon / k *gradient
    return b_elas
```

```python
n = 20
m = 20
lambda_s = np.logspace(-1,2,n)
mu_s = np.logspace(-1,2,m)
```

```python
lambda_best = 0
mu_best = 0
mse_best = 1e10
for lambda_ in lambda_s:
    b_ls = X_train.T.dot(y_train)
    b_elas = b_ls.copy()
    gradient_prev = np.zeros(b_ls.shape)
    for mu_ in mu_s:
        beta = calcBetaElastic(b_ls, b_elas, X_train, lambda_, mu_)
        mse = ((y_test - X_test.dot(beta))**2).sum()/y_test.shape[0]
        if mse_best > mse:
            mu_best = mu_
            lambda_best = lambda_
            mse_best = mse
```

```python
print("MSE best %f" %(mse_best))
print("lambda best %f" %(lambda_best))
print("mu best %f" % (mu_best))
```
